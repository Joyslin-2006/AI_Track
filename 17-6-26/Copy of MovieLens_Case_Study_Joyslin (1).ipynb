<h1><font size=6>Can we find the genres that are most liked by the users?</font></h1>


#### **Merge movie and ratings datasets**

```python
df_merge = pd.merge(ratings, movies, on="movie id")
df_merge.head()
```

#### **Calculate average rating for each genre**

```python
genre_ratings = {}

for genre in genres:
    genre_ratings[genre] = df_merge[df_merge[genre] == 1]["rating"].mean()

genre_ratings = pd.Series(genre_ratings).sort_values(ascending=False)

genre_ratings
```

#### **Visualize**

```python
genre_ratings.plot(kind='bar', figsize=(10,5))
```

Film-Noir is the highest-rated genre with an average rating of 3.92, followed by War (3.82) and Drama (3.69). This indicates that users tend to appreciate story-driven and critically acclaimed movies. On the other hand, Fantasy (3.22), Horror (3.29), and Children's movies (3.35) received comparatively lower average ratings. MovieLens can leverage this information to recommend highly-rated genres such as Film-Noir, War, and Drama to users seeking quality content.

<h1><font size=6>Can we find which movies have been most preferred by the users?</font></h1>

#### **Highest Rated Movies**

```python
movie_rating_count = df_merge.groupby("movie title")["rating"].count()

movie_rating_count.sort_values(ascending=False).head(10)
```

#### **Create average rating and count together**

```python
movie_stats = df_merge.groupby("movie title").agg({
    "rating": ["mean", "count"]
})

movie_stats.columns = ["Average Rating", "Rating Count"]

movie_stats.head()
```

#### **Filter movies with at least 50 ratings**

```python
popular_movies = movie_stats[movie_stats["Rating Count"] >= 50]

popular_movies.sort_values(
    by="Average Rating",
    ascending=False
).head(10)
```

Among movies that have received at least 50 ratings, "A Close Shave" has the highest average rating of 4.49, followed by "Schindler's List" (4.47), "The Wrong Trousers" (4.47), and "Casablanca" (4.46). These movies are not only highly rated but have also been reviewed by a substantial number of users, making them reliable indicators of audience preference. Such movies can be prioritized in recommendation systems and promotional campaigns.

<h1><font size=6>Can we extract some relationship between demographic details of the users and ratings for the movies?</font></h1>

****



#### **Merge all three datasets**

```python
df_all = pd.merge(df_merge, users, on="user id")

df_all.head()
```

#### **Average Rating by Gender**

```python
gender_ratings = df_all.groupby("gender")["rating"].mean()

gender_ratings
```

The average ratings given by male and female users are almost identical, with females giving an average rating of 3.53 and males giving an average rating of 3.53. This suggests that gender does not have a significant influence on the overall rating behavior of users in the MovieLens dataset.

```python
df_all.groupby("age")["rating"].mean().sort_values(ascending=False).head(10)
```

```python
df_all.groupby("age")["rating"].mean().sort_values().head(10)
```

Users aged 58 years give the highest average rating (3.99), followed by users aged 73 years (3.98) and 59 years (3.84). In contrast, users aged 11 years provide the lowest average rating (2.93). Overall, older users tend to give higher ratings compared to younger users, suggesting that they may have a more positive perception of the movies available on the platform.

#### **Occupation Analysis**

```python
occupation_ratings = df_all.groupby("occupation")["rating"].mean().sort_values(ascending=False)

occupation_ratings
```

Users with the occupation "none" provide the highest average rating (3.78), followed by lawyers (3.74), doctors (3.69), and educators (3.67). On the other hand, healthcare professionals provide the lowest average rating (2.90). This suggests that user occupation may influence rating behavior, with some professional groups tending to rate movies more positively than others.
