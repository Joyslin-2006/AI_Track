#### Apply Standardization
- we try to bring all the variables or features to a similar scale.
- standardization means centering the variable to zero
-  z  =(x- xmean)/std

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# fit_transform - it will transform the data according to z score
scaled =scaler.fit_transform(df_wine.drop('target',axis=1))
df_wine_scaled= pd.DataFrame(scaled,columns=df_wine.columns[:-1])
df_wine_scaled.head()
```

```python
# mean for scaled data
df_wine_scaled.mean().round(2)
```

```python
# standqard deviation
df_wine_scaled.std().round(2)
```

#### visualize after standardization

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df_wine_scaled)
plt.xticks(rotation=90)
plt.show()
```

- every features is going to contribute

##### euclidean distance in scaled data

```python
sample1 =df_wine_scaled.iloc[0][:]
sample2= df_wine_scaled.iloc[1][:]
#distance
distance=np.linalg.norm(sample1-sample2)
print(f'Distance between sample1 and sample2 is  scaled data {distance}')
```

#### indentify the largest contributor

```python
difference = abs(sample1-sample2)  # return the absolute value of the argument.
difference.sort_values(ascending=False)
```

#### Normalization -- Min Max Scaling  -- Mostly Use In Deep Learning
- Min Max Scaling scale the value in between 0 to 1
-  X scaled  = (X - X.min) / (X .max - X.min)

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()

normalized = mms.fit_transform(df_wine.drop('target',axis=1))
df_wine_normalized =pd.DataFrame(normalized, columns=df_wine.columns[:-1])
df_wine_normalized.head()
```

```python
df_wine_normalized.describe().T
```

##### Compare original  vs Standardized vs Normalized

```python
fig,ax =plt.subplots(1,3,figsize=(18,9))

sns.boxplot(data=df_wine[['alcohol','magnesium','proline']],ax=ax[0])
ax[0].set_title('Original Data')

sns.boxplot(data=df_wine_scaled[['alcohol','magnesium','proline']],ax=ax[1])
ax[1].set_title('Standardized Data')


sns.boxplot(data=df_wine_normalized[['alcohol','magnesium','proline']],ax=ax[2])
ax[2].set_title('Normalized Data')
```

##### Robust Scaler

- it used to scale the features between median and quantiles.
- scaling using meadian and quantiles consist of substracting the median with all observations.
- then dividing by IQR.it difference between 75th and 25th quantile
- X_robust_scaled =  (X - Xmedian)/IQR

```python
from sklearn.preprocessing import RobustScaler
robustScaler = RobustScaler()
df_wine_robustscaled =pd.DataFrame(robustScaler.fit_transform(df_wine.drop('target',axis=1)),
                                   columns=df_wine.columns[:-1])
df_wine_robustscaled.head()
```

```python
fig,ax =plt.subplots(1,4,figsize=(18,9))

sns.boxplot(data=df_wine[['alcohol','magnesium','proline']],ax=ax[0])
ax[0].set_title('Original Data')

sns.boxplot(data=df_wine_scaled[['alcohol','magnesium','proline']],ax=ax[1])
ax[1].set_title('Standardized Data')


sns.boxplot(data=df_wine_normalized[['alcohol','magnesium','proline']],ax=ax[2])
ax[2].set_title('Normalized Data')

sns.boxplot(data=df_wine_robustscaled[['alcohol','magnesium','proline']],ax=ax[3])
ax[3].set_title('Robust Scaled')
```

#### Q-Q plot - is used to check whether a feature is guassian or normal distributed

```python
import scipy.stats as stat
import pylab
```

```python
def plot_data(df,feature):
  plt.figure(figsize=(12,6))
  plt.subplot(1,2,1)
  df[feature].hist()
  plt.subplot(1,2,2)
  stat.probplot(df[feature],dist='norm',plot=pylab)
  plt.title ("Q-q plot "+feature)
  plt.show()
```

```python
df_wine = pd.DataFrame(data_wine.data, columns=data_wine.feature_names)
df_wine.head()
```

```python
plot_data(df_wine,'alcohol')
```

```python
for column in df_wine.columns:
  plot_data(df_wine,column)
```
