## Binomial Distribution

Problem statement
80% of all the visitors to Lavista Museum end up buying souvenirs from the souvenir shop at the Museum. On the coming Sunday, if a random sample of 10 visitors is picked:

1. Find the probability that every visitor will end up buying from the souvenir shop  

2. Find the probability that a maximum of 7 visitors will buy souvenirs from the souvenir shop

Let's check first whether we satisfy the assumptions of the binomial distribution.   There are only two possible outcomes (success or failure) for each trial

– A visitor will buy souvenirs from the souvenir shop or not (yes or no).  Number of trials (n) is fixed
- There are 10 visitors in the sample.  Each trial is independent of the other trials
- It is reasonable to assume that the buying activity of visitors is independent.  The probability of success (p) is the same for each trial
- The probability of success for each visitor is 0.8.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats   # this lib contains large number of probability distribution as well statistical functions.
```

### Let eastimate the probability distribution of visitors

```python
# sample size
n=10
```

```python
# declare p which represent probabilty of success  :  probability a visitor will buy a souvenir
p=0.80
```

```python
# different possible number of visitor
k = np.arange(0,11)
```

```python
# for binomial distribution function import
from scipy.stats import binom
```

```python
# to generate binomial probability distribution   k = scalar , array of different posibile no of visitors , n  sample size , p prob of sucess
bibinomial =binom.pmf(k=k,n=n ,p=p)
```

```python
bibinomial
```

```python
plt.bar(k,bibinomial)
plt.grid()
plt.title("Binomial : n =%i , p = %.2f" %(n,p), fontsize=12)
plt.xlabel("Number of success")
plt.ylabel("Probability of success")
```

```python
bibinomial[10]
```

#### the above probability shows the probability of exactly 10 sucess is ~0.11

#### CDF - cummulative probabilty
         - of a random variable X is the probability that X will take the value
         less than equal to  x .   Fx(x) = P(X <=  x )
         P(X<=7 )

```python
barl=plt.bar(k,bibinomial)
plt.grid()
plt.title("Binomial : n =%i , p = %.2f" %(n,p), fontsize=12)
plt.xlabel("Number of success")
plt.ylabel("Probability of success")
for i in range(0,8):
  barl[i].set_color('r')
```

```python
# calculate cdf
binom.cdf(k=7,n=n,p=p)
```

### Conclusion

-  there 10.74% chances that every visitor will end up buying a souvenirs from the shop.

- there 32.22% chances that a max of  7 out of 10 visitor will end up buying a souvenirs from the shop
