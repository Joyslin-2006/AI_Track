```python
print('sum of all elements of array',np.sum(mat))   # axis = 0 is vertical or columns  . axis =1 rows or acrros
print('max of all elements of array',np.max(mat))
print('min of all elements of array',np.min(mat))
print('cumsum of all elements of array',np.cumsum(mat))
```

```python
print('sum of rows elements of array ',np.sum(mat,axis=1))
print('max of row elements of array',np.max(mat,axis=1))
print('min of row elements of array',np.min(mat,axis=1))
print('cumsum of row elements of array',np.cumsum(mat,axis=1))
```

```python
print('sum of column elements of array ',np.sum(mat,axis=0))
print('max of column elements of array',np.max(mat,axis=0))
print('min of column elements of array',np.min(mat,axis=0))
print('cumsum of column elements of array',np.cumsum(mat,axis=0))
print('mean of column elements of array',np.mean(mat,axis=0))
```
